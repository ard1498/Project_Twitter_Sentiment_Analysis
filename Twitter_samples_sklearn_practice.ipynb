```
import json
neg_tweets = []
for line in twitter_samples.open(files[0]):
  neg_tweets.append(json.loads(line))
len(neg_tweets)
```

```
pos_tweets = []
for line in twitter_samples.open(files[0]):
  pos_tweets.append(json.loads(line))
len(pos_tweets)
```

```
import nltk
nltk.download('punkt')
```

```
neg_tweets = [tweets['text'] for tweets in neg_tweets]
pos_tweets = [tweets['text'] for tweets in pos_tweets]
len(neg_tweets),len(pos_tweets),neg_tweets[0],pos_tweets[0]
```

```
from nltk.tokenize import word_tokenize
neg_docs = [word_tokenize(sen) for sen in neg_tweets]
pos_docs = [word_tokenize(sen) for sen in pos_tweets]
len(neg_docs),neg_docs[0],len(pos_docs),pos_docs[0]
```
